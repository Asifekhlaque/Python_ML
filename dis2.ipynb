```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"/content/Experiment_2 - Sheet2.csv")
df
```

```python
input=df.drop(['salary_100k'], axis='columns')
target=df['salary_100k']
```

```python
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
input['company_n']=le_company.fit_transform(input['company'])
input['job_n']=le_job.fit_transform(input['job'])
input['degree_n']=le_degree.fit_transform(input['degree'])
```

```python
model=tree.DecisionTreeClassifier()
model.fit(input[['company_n','job_n','degree_n']],target)
```

```python
model.predict([[2,0,0]])
```

```python
acc=model.score(input[['company_n','job_n','degree_n']],target)
acc
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(model,filled=True)
plt.show()
```
